```python
#creating dataframe 1
df1=pd.DataFrame({
    "city":['mumbai','hyd','chennai','pune','banglore'],
    "temperature":[27,32,23,36,46],
    })
```

```python
df1
```

```python
#creating dataframe 2
df2=pd.DataFrame({
    "city":['new york','chicago','hyd','kolkata','chennai','orlando'],
    "humidity":[68,73,66,90,22,46]
})
```

```python
df2
```

```python
#outer merge method_union of 2 dataframes
df=pd.merge(df1,df2,how="outer")
df
```

```python
#outer merge method_union of 2 dataframes
df=pd.merge(df1,df2,on="city",how="outer")
```

```python
df
```

```python
#inner merge method -intersection of 2 dataframes
df=pd.merge(df1,df2,on="city",how="inner")
```

```python
df
```

```python
#left merge method-use keys from left frame i.e., from first dataframe
df=pd.merge(df1,df2,on="city",how="left")
```

```python
df
```

```python
#right merge method-use keys from right frame i.e., from second dataframe
df=pd.merge(df1,df2,on="city",how="right")
```

```python
df
```

```python
#by default inner join
df=pd.merge(df1,df2,on="city")
```

```python
df
```

```python
#to check wheather the NAN values are from left or right dataframe
#we set indicator to True
#by default indicator is false
df=pd.merge(df1,df2,on="city",how='outer',indicator=True)
df
```

# Dataframe with same attributes(column names)

```python
#suppose the dataframes has same columns
#here city,humidity,temperature is same from dataframe 2
#creating dataframe 1
df1=pd.DataFrame({
    "city":['mumbai','hyd','chennai','pune','banglore'],
    "temperature":[27,32,23,36,46],
    "humidity":[80,60,78,66,22]
})
```

```python
df1
```

```python
#suppose the dataframes has same columns
#here city,humidity,temperature is same from dataframe 1
#creating dataframe 2
df2=pd.DataFrame({
    "city":['new york','chicago','hyd','kolkata','chennai','orlando'],
    "temperature":[36,22,32,22,78,14],
    "humidity":[68,73,66,90,22,46]
})
```

```python
df2
```

```python
#left merge method-use keys from left frame i.e., from first dataframe
df=pd.merge(df1,df2,on="city",how="left")
df
```

```python
#right merge method-use keys from right frame i.e., from second dataframe
df=pd.merge(df1,df2,on="city",how="right")
df
```

```python
#outer merge method_union of 2 dataframes
df=pd.merge(df1,df2,on="city",how="outer")
df
```

```python
#inner merge method -intersection of 2 dataframes
df=pd.merge(df1,df2,on="city",how="inner")
df
```

```python
#by default inner join
df=pd.merge(df1,df2,on="city")
df
```

```python
#to check wheather the NAN values are from left or right dataframe
#we set indicator to True
#by default indicator is false
df=pd.merge(df1,df2,on="city",how='outer',indicator=True)
df
```

```python
#on merging these two dataframes 
#we get column attributes as humidity_x,temperature_x,humidity_y,temperature_y
#indicating x refers to dataframe_1 and y refers to dataframe_2
#suppose u  want to change this names, we use a parameter called suffixes as below
df=pd.merge(df1,df2,on="city",how="outer",suffixes=("_left","_right"))
```

```python
df
```

```python
df=pd.merge(df1,df2,on="city",how="outer",suffixes=("_left","_right"),indicator=True)
```
