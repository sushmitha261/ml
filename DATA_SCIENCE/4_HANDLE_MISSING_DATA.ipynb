# 1.By fillna

```python

new_df=df.fillna(0)
```

```python
new_df
```

```python
#sometimes filing NaN with zeros may not be the correct 
#As observed in event column NaN are filled by 0 ,which is meaningless
#Thus,pandas support to fill NaN for specified columns using dictionary as below
#DataFrame.fillna({key:values,key:values,..})
new_df=df.fillna(
    {'temperature':0,'windspeed':0,'event':'No event'}
)
```

```python
new_df
```

```python
#As  we are not aware of the temperatures,windspeed on some days
#we have filled it with 0.But this may not be appropiate
#Thus we can carry-out these values either by mean value or 
#by previous days temperature or next days temperature
```

```python
#fillna by previous data(all NaN values are filled by its previous data)
#method=ffill means forward fill
new_df=df.fillna(method='ffill')
```

```python
new_df
```

```python
#fillna by its next data(all NaN values are filled by its next data)
#method=ffill means backward fill
new_df=df.fillna(method='bfill')
```

```python
new_df
```

```python
#by using the parameter axis filling the NaN in the column wise 

new_df=df.fillna(method='bfill',axis='columns')
```

```python
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)
```

```python
new_df
```
