```python
import pandas as pd
```

```python
#When we concatenated our DataFrames we simply added them to each other,
#stacking them either vertically or side by side.
```

```python
#The main difference: pd.concat works on both axes.
#The other difference, is pd.concat has innerdefault and outer joins only,
#while pd.DataFrame.merge () has left, right, outer, innerdefault joins.
```

# example_1

```python
india_weather=({
    "city":['mumbai','hyd','banglore'],
    "temperature":[27,32,23],
    "humidity":[80,60,78]
})
```

```python
india_weather
```

```python
#creating dataframe 1
india_weather=pd.DataFrame({
    "city":['mumbai','hyd','banglore'],
    "temperature":[27,32,23],
    "humidity":[80,60,78]
})
```

```python
india_weather
```

```python
#creating dataframe 2
us_weather=pd.DataFrame({
    "city":['new york','chicago','orlando'],
    "temperature":[36,22,14],
    "humidity":[68,70,55]
})
```

```python
us_weather
```

```python
#concating the dataframes
df=pd.concat([india_weather,us_weather])
```

```python
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
```

```python
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True,keys=["india","us"])
```

```python
df
```

```python
#to partition the dataframe
df=pd.concat([india_weather,us_weather],keys=["india","us"])
```

```python
df
```

```python
#to retrive india_weather
df.loc["india"]
```

```python
#to retrive us_weather
df.loc["us"]
```

# example_2

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","hyd","banglore"],
    "temperature":[54,56,44]
})
```

```python
temperature_df
```

```python
windspeed_df=pd.DataFrame({
    "city":["mumbai","hyd","banglore"],
    "windspeed":[5,6,4]
})
```

```python
windspeed_df
```

```python
#by deafault they append by rows
pd.concat([temperature_df,windspeed_df])
```

```python
#axis=1 indiates that dataframe2 appends as columns
pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
#axis=0 indiates that dataframe2 appends as columns
pd.concat([temperature_df,windspeed_df],axis=0)
```

# example_3

```python
#suppose the dataframes are incorrect i.e., not inorder
#dataframe 1

temperature_df=pd.DataFrame({
    "city":["mumbai","hyd","banglore"],
    "temperature":[54,56,44]
})

#dataframe 2

windspeed_df=pd.DataFrame({
    "city":["banglore","mumbai"],
    "windspeed":[8,3]
})
```

```python
temperature_df
```

```python
windspeed_df
```

```python
#axis=0 indiates that dataframe2 appends as columns
pd.concat([temperature_df,windspeed_df],axis=0)
```

```python
#axis=1 indiates that dataframe2 appends as columns
pd.concat([temperature_df,windspeed_df],axis=1)
```

```python
#suppose the dataframes are incorrect i.e., not inorder
#we pass index to make sure that they are in order
#dataframe 1

temperature_df=pd.DataFrame({
    "city":["mumbai","hyd","banglore"],
    "temperature":[54,56,44]
},index=[0,1,2])

#dataframe 2

windspeed_df=pd.DataFrame({
    "city":["banglore","mumbai"],
    "windspeed":[8,3]
},index=[2,0])
```

```python
#axis=0 indiates that dataframe2 appends as columns
pd.concat([temperature_df,windspeed_df],axis=0)
```

```python
#axis=1 indiates that dataframe2 appends as rows
pd.concat([temperature_df,windspeed_df],axis=1)
```
